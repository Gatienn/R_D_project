```python
import methode_derivee as md
import numpy as np
import matplotlib.pyplot as plt
```

```python
score_max, score_min, score_vrais_seuils, score_moy = md.score_quantile(N=100, step = 0.1, nb_points=5, noise = 0.0, table = md.table_derivees)
```

```python
plt.hist(score_max, bins = 10, label = 'score max', alpha = 0.5)
plt.hist(score_vrais_seuils, bins = 10, label = 'score vrais seuils', alpha = 0.5)
plt.hist(score_moy, bins = 10, label = 'score moyen', alpha = 0.5)
#plt.hist(score_min, bins = 10, label = 'score min', alpha = 0.5)
plt.legend(loc = 'upper right')
plt.xlabel('Score')
plt.ylabel('Nombre de simulations')
plt.show()
```
